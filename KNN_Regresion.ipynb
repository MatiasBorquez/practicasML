```python
from sklearn import datasets
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV, train_test_split
import pandas as pd
```

```python
diabetes = datasets.load_diabetes()
```

```python
diabetes
```

```python
regressor = KNeighborsRegressor()
X_train, X_test, y_train, y_test = train_test_split(diabetes.data, diabetes.target, test_size = 0.1)
```

```python
param_grid = {"n_neighbors": range(1,51), "weights": ["uniform", "distance"]}
```

```python
gs = GridSearchCV(estimator= regressor, param_grid= param_grid, scoring= None, cv= None)
```

```python
gs.fit(X_train, y_train)
```

```python
pd.DataFrame(gs.cv_results_)
```

```python
gs.best_params_
```

```python
y_pred = gs.best_estimator_.predict(X_test)
```

```python
pd.DataFrame({"y_test": y_test, "y_pred": y_pred})
```

```python
best_model = gs.best_estimator_
```

```python
best_model.score(X_test, y_test)
```

```python
best_model.get_params()
```
